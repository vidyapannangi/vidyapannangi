```python
# Importing the 're' module from the 'regex' library
import regex as re

# Defining a function 'tokenize' that uses regular expressions to extract words from the given text
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
# Test text containing a mix of words, numbers, and special characters
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"

# Tokenizing the test text using the previously defined 'tokenize' function
testtokens = tokenize(testtext)

# Printing the extracted tokens
print(testtokens)
```

```python
# Importing nltk library and downloading the stopwords dataset
import nltk
nltk.download('stopwords')

# Creating a set of English stopwords
stopwords = set(nltk.corpus.stopwords.words('english'))

# Printing the set of stopwords
print(stopwords)
```

```python
# Function to remove stopwords from a list of tokens
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
```

```python
# Removing stopwords from the list of tokens
cleantokens = remove_stop(testtokens)
print(cleantokens)
```

```python
# Defining a text processing pipeline
pipeline = [str.lower, tokenize, remove_stop]
```

```python
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
# Apply the text processing pipeline to the 'text' column in the DataFrame
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

```python
# Count the number of tokens in each row and create a new column 'num_tokens'
df['num_tokens'] = df['tokens'].map(len)

# Display the last few rows of the DataFrame
df.tail()
```

```python
# Import the Counter class from the collections module
from collections import Counter
```

```python
# Apply the Counter function to count the occurrences of each token
tokens = tokenize("She likes my cats and my cats like my sofa.")
testcount = Counter(tokens)
print(testcount)
```

```python
# Update the counter with additional tokens
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

```python
# Create a Counter to count the frequency of tokens in the entire DataFrame
counter = Counter()

# Update the counter with tokens from each row in the 'tokens' column
df['tokens'].map(counter.update)

# Display the most common 5 tokens
print(counter.most_common(5))
```

```python
# Create a DataFrame from the Counter and filter tokens with frequency greater than or equal to 1
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
# Sort the DataFrame by frequency in descending order and display the top 10 tokens
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
# Display descriptive statistics for the frequency DataFrame
freq_df.describe()
```

```python
# Create a horizontal bar plot for the top 15 words
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```

```python
# Import necessary libraries
from wordcloud import WordCloud
from matplotlib import pyplot as plt
```

```python
# Extracting text for the specified conditions
text = df.query("year==1989 and country=='IND'")['text'].values[0]
# Creating and displaying a Word Cloud
plt.figure(figsize=(14, 7))
wc = WordCloud(max_words=100, stopwords=stopwords)
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

```python
# Specify conditions
conditions = (df['year'] == 1989) & (df['country'] == 'IND')

# Check if there is data for the specified conditions
if df[conditions].empty:
    print("No data found for the given conditions.")
else:
    # Extract text for the specified conditions
    text = df[conditions]['text'].values[0]

    # Create and generate the word cloud
    wc = WordCloud(background_color='white', width=300, height=300, max_words=100, margin=2)
    wc.generate_from_text(text)

    # Display the word cloud
    plt.figure(figsize=(8, 8), facecolor=None)
    plt.imshow(wc)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.show()
```
