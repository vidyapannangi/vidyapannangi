# Loading The DataSet

```python
import pandas as pd #Iporting pandas as pd
file = "un-general-debates-blueprint.csv" #File name
df = pd.read_csv(file)
df.head(20) # Printing top 20 rows
```

# Importing nltk Package

```python
import nltk
nltk.download('stopwords') # Downloading Stop Words
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords) # Printing Stop Words
```

# Pre Processing Pipeline

```python
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text 
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

# Counting The Tokens

```python
from collections import Counter
counter = Counter()
df['tokens'].map(counter.update)
freq_df = pd.DataFrame.from_dict(counter, orient='index',columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token’
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```
